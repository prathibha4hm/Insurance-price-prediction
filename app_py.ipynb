```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import streamlit as st
from sklearn.preprocessing import MinMaxScaler

#load model
model = pickle.load(open('model_gb.pkl', 'rb'))

#scaling
scaler = MinMaxScaler()

#give title
st.title("Insurance premium price prediction app")

age = st.numer_input('Age',min_value=1,max_value = 100,value=25)
gender = st.selectbox('Gender',('male','female'))
bmi = st.number_input('BMI',min_value=10.0,max_val=100.0,value=30.0)
smoker = st.selectbox('Smoker',('yes','no'))
children = st.selectbox('Number of children',min_value=1,max_value=10,value=2)
region = st.selectbox('Region',('southeast','southwest','northwest','northeast'))

#encoding

#smoker
Smoker = 1 if smoker == 'yes' else 0

#gender
sex_female = 1 if gender == 'female' else 0
sex_male = 1 if gender == 'male' else 0

#region
region_dict = {'southeast':0,'southwest':1,'northwest':2,'northeast':3}
region_code = region_dict[region]

#dataframe
input_features = pd.DataFrame({
    'age':[age],
    'sex_female':[sex_female],
    'sex_male':[sex_male],
    'bmi':[bmi],
    'children':[children],
    'smoker':[Smoker],
    'region_code':[region_code]
})

input_features[['age','bmi']] = scaler.fit_transform(input_features[['age','bmi']])

#predictions
if st.button('Predict'):
  predictions = model.predict(input_features)
  output = round(np.exp(predictions[0]),2)
  st.success(f'Price Prediction:${output}')
```
